```python
import pandas
import os
```

```python
HOME_DIR = os.environ['HOME']

elo_data = HOME_DIR + "/repos/nfl-elo-game/data/nfl_games_2010.csv"
df_elo = pandas.read_csv(elo_data)
print len(df_elo)

# remove playoffs 
playoff_filter = df_elo.playoff == 1
df_elo = df_elo[~playoff_filter]

print len(df_elo)
```

```python

lookup_file = HOME_DIR + "/repos/nfl-elo-game/data/team_lookup.csv"
df_lookup = pandas.read_csv(lookup_file)
df_lookup
```

```python
logreg_data = HOME_DIR + "/repos/nfl-elo-game/data/logreg_2010.csv"
df_logreg = pandas.read_csv(logreg_data)
df_logreg
```

```python
df_merge1 = df_elo.merge(df_lookup, left_on = 'team1', right_on = 'team').merge(df_lookup, left_on = 'team2', right_on = 'team', suffixes=("1","2"))
```

```python
df_final = df_merge1.merge(df_logreg, left_on = ['name1','name2','season'], right_on =['Home Team', 'Visitor','season']).sort_values('week')
```

```python
df_final.eloScore.sum(), df_logreg.eloScore.sum()
```

```python
df_final.pivot_table(index='season', values="eloScore", aggfunc=len)
```

```python
df_final.pivot_table(index='season', values="eloScore", aggfunc=sum)
```

```python
away_filter = df_final.Visitor == df_final.favorite
df_final.loc[~away_filter,'my_prob1'] = df_final.predict_proba
df_final.loc[away_filter,'my_prob1'] = 1 - df_final.predict_proba
```

```python
new_elo_data = HOME_DIR + "/repos/nfl-elo-game/data/new_nfl_games_2010.csv"
df_final.to_csv(new_elo_data, index=False)
```
