```python
import pandas
import os
```

```python
ROOT_DIR = os.environ["MLNFL_ROOT"]
DATA_DIR = ROOT_DIR + "/data/lines"

team_file = DATA_DIR + "/nflteams2019.csv"
lines_file = DATA_DIR + "/nfl2019lines.csv"
```

```python
df_teams = pandas.read_csv(team_file)
df_lines = pandas.read_csv(lines_file)
```

```python
df_lines.head()
```

```python
df_teams.head()
```

```python
lines_cols = ['game_id','home_team', 'away_team', 'week', 'season']
teams_cols = ['team', 'abbr']
df_merge = df_lines[lines_cols].merge(df_teams[teams_cols], left_on="home_team", right_on="abbr")
df_games = df_merge.merge(df_teams[teams_cols],left_on="away_team", right_on="abbr", suffixes=("_home", "_away")).sort_values('game_id')
df_games
```

```python
df_games['date'] = pandas.to_datetime(df_games['game_id'].astype(str).str[:-2])
```

```python
use_cols = ['date', 'team_away', 'team_home', 'season', 'week']

df_games[use_cols].to_clipboard()
```
