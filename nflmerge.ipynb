```python
# define source and data directories
import os
homeDir = os.environ['HOME'] + '/'

codeDir = homeDir + 'repos/mlkaggle/nfl/'
codeDir = homeDir + 'sandbox/nfl/'
dataRoot = codeDir + "data/" 
os.chdir(codeDir)

dataRoot
```

```python
import pandas
import csv


```

```python
allSeasons = range(1978,2014)

outfile = dataRoot + "lines/nflAllLines.csv"

dfAll = None
for ss in allSeasons:
    datafile = dataRoot + ("lines/nfl%slines.csv" % ss)
    # print datafile
    df = pandas.read_csv(datafile)
    df['season'] = ss
    if dfAll is None:
        dfAll = df
    else:
        dfAll = dfAll.append(df)


dfAll.to_csv(outfile,index=False)
```
