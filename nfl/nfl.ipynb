```python
# define source and data directories
import os
homeDir = os.environ['HOME'] + '/'

codeDir = homeDir + 'repos/mlkaggle/nfl/'
codeDir = homeDir + 'sandbox/nfl/'
dataRoot = codeDir + "data/" 
os.chdir(codeDir)

print codeDir
dataRoot
```

```python
# lookup files
import lookup
lookupFiles = { 'teams' : {'file': 'nflTeams.csv' },
                'seasons' : {'file': 'seasons.csv' },
               }

lookupDir = dataRoot+'lookup/'
olookups = lookup.Lookup(lookupDir, lookupFiles)

# print out for diagnostics
#olookups.ldict['seasons'].keys()
```

```python
### NEW multi-season training
import data
reload(data)

# need to be arrays
trainYears = range(2008,2013) 
testYear = [2013]

# training data set - includes one extra year for prev yr record
seasons = np.array(trainYears)
print "training seasons >> ", seasons
dfAllGames = data.readGamesAll(dataRoot,seasons)

# compile season record for all teams
dfAllTeams = data.seasonRecord(dfAllGames,olookups)

# apply season records and compute other fields for all games
dfAllGames = data.processGames(dfAllGames, dfAllTeams, olookups)
# remove extra year of data 
dfAllGames = dfAllGames[dfAllGames.season.isin(seasons)]

# define indep variables for logistic regression
features = ['favoredRecord','underdogRecord',
            'prevFavoredRecord','prevUnderdogRecord',
            'gameWeek',
            'absLine',
            'divisionGame',
            'favoredHomeGame',
            ]
# run the logistic regression
logreg = data.runML(dfAllGames,features)

# use different test set 
seasonTest = np.array(testYear) # should be only one year
print "results for >> ", seasonTest
dfGamesTest = data.readGamesAll(dataRoot,seasonTest)
dfTeamsTest = data.seasonRecord(dfGamesTest,olookups)
dfGamesTest = data.processGames(dfGamesTest, dfTeamsTest, olookups)
dfGamesTest = dfGamesTest[dfGamesTest.season.isin(seasonTest)]

# apply results of logistic regression to the test set
dfPredict = data.predictGames(dfGamesTest,logreg,features)

# apply ranking logic and determine scoring outcomes for league
dfAll = data.rankGames(dfPredict,olookups,seasonTest[0])
```
